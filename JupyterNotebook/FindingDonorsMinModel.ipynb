```python
# Import libraries necessary for this project
import numpy as np
import pandas as pd
from time import time
from IPython.display import display # Allows the use of display() for DataFrames

# Import supplementary visualization code visuals.py
import visuals as vs

# Pretty display for notebooks
%matplotlib inline

# Load the Census dataset
data = pd.read_csv("census.csv",usecols=['age','hours-per-week','capital-gain','education-num','relationship','marital-status','income'])

# Success - Display the first record
display(data.head(n=1))
```

```python
#From our research we found that we need the following minimal features 
# age  , hours per week , capital gain , education num ,relationship , Marritial status 
# 

#numerical = ['age','hours-per-week','capital-gain','education-num']
```

```python
log_transform = ['capital-gain']
```

```python
numerical = ['age','hours-per-week','capital-gain','education-num']
```

```python
income_raw = data['income']
features_raw = data.drop('income', axis = 1)

```

```python
features_raw.head()
```

```python
features_raw[log_transform] = features_raw[log_transform].apply(lambda x: np.log(x + 1))
```

```python
features_raw.head()
```

```python
#Normalizing numerical features 
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() # default=(0, 1)
features_raw[numerical] = scaler.fit_transform(features_raw[numerical])

# Show an example of a record with scaling applied
display(features_raw.head(n = 5))
```

```python
features_final = pd.get_dummies(features_raw)
features_final.head()
income = income_raw.replace({'<=50K':0, '>50K':1})
#ConcatDataset = features_final.join(income)
```

```python
features_final.to_csv("test.csv",index=False)
```

```python
features_final.head()
```

```python
data = pd.read_csv("test.csv")
```

```python
data.head()
```

```python
# Import train_test_split

#from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split

# Split the 'features' and 'income' data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, 
                                                    income, 
                                                    test_size = 0.2, 
                                                    random_state = 0)

# Show the results of the split
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(criterion='friedman_mse', init=None,
              learning_rate=0.2, loss='deviance', max_depth=8,
              max_features=None, max_leaf_nodes=None,
              min_impurity_decrease=0.0, min_impurity_split=None,
              min_samples_leaf=1, min_samples_split=2,
              min_weight_fraction_leaf=0.0, n_estimators=15,
              presort='auto', random_state=1, subsample=1.0, verbose=0,
              warm_start=False)
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import fbeta_score
predictions = clf.predict(X_test)
print("Accuracy on testing data: {:.4f}".format(accuracy_score(y_test, predictions)))
print("F-score on testing data: {:.4f}".format(fbeta_score(y_test, predictions, beta = 0.5)))
```
